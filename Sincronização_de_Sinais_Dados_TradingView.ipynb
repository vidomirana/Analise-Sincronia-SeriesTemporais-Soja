# Sincronização

```python
from scipy.signal import hilbert
```

```python
sojagrao = hilbert(curvaSOY.preco)
sojaoleo = hilbert(curvaOIL.preco)
sojafarelo = hilbert(curvaMEAL.preco)
```

## FARELO - OLEO

```python
angulofarelo = np.angle(sojafarelo,deg=False)
angulooleo = np.angle(sojaoleo,deg=False)
f,ax = plt.subplots(3,1,figsize=(20,12),sharex=True)
ax[0].plot(curvaMEAL.preco,color='r',label='Farelo de Soja')
ax[0].plot(curvaOIL,color='b',label='Óleo de Soja')
ax[0].legend(bbox_to_anchor=(0., 1.02, 1., .102),ncol=2)
ax[0].set(ylim=[0,450], xlim=[0,len(curvaMEAL)], title='Preços do farelo e óleo de soja')
ax[1].plot(angulofarelo,color='r')
ax[1].plot(angulooleo,color='b')
ax[1].set(ylabel='Angulo',title='Angulo de cada intervalo de tempo')
sincronia = 1-np.sin(np.abs(angulofarelo-angulooleo)/2)
#sincronia = angulograo - angulooleo
ax[2].plot(sincronia)
ax[2].set(ylim=[-1,1.1],xlim=[0,len(curvaMEAL)],title='Sincronização instantânea',xlabel='Tempo',ylabel='Sincronização de Fase')
plt.tight_layout()
plt.show()
```
